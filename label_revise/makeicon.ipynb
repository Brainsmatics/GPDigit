```python
import PythonMagick
import base64
```

```python
img = PythonMagick.Image("refresh.jpg")
```

```python
img.sample('56x56')
```

```python
img.write("logo.ico")
```

```python
open_icon = open("logo.ico","rb")
b64str = base64.b64encode(open_icon.read())
open_icon.close()
write_data = "img = %s" % b64str
f = open("icon.py","w+")
f.write(write_data)
f.close()
```
